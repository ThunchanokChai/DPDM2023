#####filler out outlier
เนื่องจากข้อมูลของ China และ United Kingdom มี outlier ดังนั้นต้องลบค่า outlier ออก

###### filler out outlier : China

````python
boxplot_China['whiskers'][0].get_ydata().min()   #find the value min of whiskers
````

````python
boxplot_China['whiskers'][1].get_ydata().max()   #find the value min of whiskers
````

````python
Data_HW3[(Data_HW3['China'] >=  boxplot_China['whiskers'][0].get_ydata().min() )
& (Data_HW3['China'] <=  boxplot_China['whiskers'][1].get_ydata().max() )]
````

````python
Data_HW3_cut_outlier_China=Data_HW3[(Data_HW3['China'] >=  boxplot_China['whiskers'][0].get_ydata().min() )
& (Data_HW3['China'] <=  boxplot_China['whiskers'][1].get_ydata().max() )]
````

###### filler out outlier : United Kingdom

````python
boxplot_United_Kingdom['whiskers'][0].get_ydata().min()   #find the value min of whiskers
````

````python
boxplot_United_Kingdom['whiskers'][1].get_ydata().max()   #find the value min of whiskers
````

````python
Data_HW3_cut_outlier_China[(Data_HW3_cut_outlier_China['United Kingdom'] >=  boxplot_United_Kingdom['whiskers'][0].get_ydata().min() )
& (Data_HW3_cut_outlier_China['United Kingdom'] <=  boxplot_United_Kingdom['whiskers'][1].get_ydata().max() )]
````

````python
#จะได้ตารางข้อมูลใหม่ที่ตัด outlier ออกทั้งหมดแล้วดังนี้
Data_HW3_New=Data_HW3_cut_outlier_China[(Data_HW3_cut_outlier_China['United Kingdom'] >=  boxplot_United_Kingdom['whiskers'][0].get_ydata().min() )
& (Data_HW3_cut_outlier_China['United Kingdom'] <=  boxplot_United_Kingdom['whiskers'][1].get_ydata().max() )]
````

````python
Data_HW3_New
````

####Plot

#####Time Series

````python
plt.plot(range(len(Data_HW3_New)),Data_HW3_New['Canada'],'o--b',label='Canada')
plt.plot(range(len(Data_HW3_New)),Data_HW3_New['China'],'o--y',label='China')
plt.plot(range(len(Data_HW3_New)),Data_HW3_New['Switzerland'],'o--r',label='Switzerland')
plt.plot(range(len(Data_HW3_New)),Data_HW3_New['United Kingdom'],'o--g',label='United Kingdom')
plt.plot(range(len(Data_HW3_New)),Data_HW3_New['United States'],'o--k',label='United States')
plt.title('New COVID-19 Canada VS Switzerland') #create name
plt.xlabel('Date')
plt.ylabel('New case COVID-19')
plt.legend();
````

#####Stacked barchart

````python
plt.bar(range(len(Data_HW3_New)), Data_HW3_New['Canada'],color='b',label='Canada')
plt.bar(range(len(Data_HW3_New)), Data_HW3_New['China'],bottom=Data_HW3_New['Canada'],color='y',label='China')
plt.bar(range(len(Data_HW3_New)), Data_HW3_New['Switzerland'],bottom=Data_HW3_New['China']+Data_HW3_New['Canada'],color='r',label='Switzerland')
plt.bar(range(len(Data_HW3_New)), Data_HW3_New['United Kingdom'],bottom=Data_HW3_New['Switzerland']+Data_HW3_New['China']+Data_HW3_New['Canada'],color='g',label='United Kingdom')
plt.bar(range(len(Data_HW3_New)), Data_HW3_New['United States'],bottom=Data_HW3_New['United Kingdom']+Data_HW3_New['Switzerland']+Data_HW3_New['China']+Data_HW3_New['Canada'],color='k',label='United States')
plt.grid(True,axis='y')
plt.legend();
````
